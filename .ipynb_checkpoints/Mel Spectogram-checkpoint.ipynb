```python
import numpy as np, scipy as sp
import matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
from scipy import signal
from scipy.fft import fft
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual

# %matplotlib widget
%matplotlib inline
```

```python
# Load Audio
audio, sr = librosa.load("audio/cello.wav")
```

```python
filterBank = librosa.filters.mel(n_fft=2**11, sr=22050, n_mels=13)
filterBank.shape
```

```python
plt.figure(figsize=(15, 7.5))
librosa.display.specshow(filterBank, sr=sr, x_axis='linear', y_axis='mel')
plt.colorbar(format="%+.4f")
plt.title('Banco de Filtros')
plt.tight_layout()
plt.show()
```
